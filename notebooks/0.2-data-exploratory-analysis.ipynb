# Dataset exploratory analysis
Understanding the data is essential part of creating models that perform well. We can understand data through data analysis.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../data/interim/EEA-SK-Ba.csv')
df.head()
```

```python
print(f'Number of rows: {len(df)}')
print(f'Number of rows containing NaN: {len(df[df.isna().any(axis=1)])}')
```

```python
df.dtypes
```

```python
df.describe().transpose()
```

```python
df.plot(subplots=True, title='Dáta z rokov 2018 až 2021')
```

```python
df[:720].plot(subplots=True, title='Január 2018')
```

```python
df[:7*24].plot(subplots=True, title='Prvý týždeň januára 2018')
```

```python
ax = sns.boxplot(data=df[['PM10 Concentration', 'PM2.5 Concentration', 'CO Concentration', 'NO2 Concentration', 'SO2 Concentration', 'C6H6 Concentration']])
ax.set_xticklabels(ax.get_xticklabels(),rotation=70)
```

```python
df[['PM10 Concentration', 'PM2.5 Concentration', 'CO Concentration', 'NO2 Concentration', 'SO2 Concentration', 'C6H6 Concentration']].corr(method='pearson')
```

```python
sns.heatmap(data=df[['PM10 Concentration', 'PM2.5 Concentration', 'CO Concentration', 'NO2 Concentration', 'SO2 Concentration', 'C6H6 Concentration']].corr(method='pearson'), annot=True)
```

```python
sns.pairplot(data=df[['PM10 Concentration', 'PM2.5 Concentration', 'CO Concentration', 'NO2 Concentration', 'SO2 Concentration', 'C6H6 Concentration']])
```
